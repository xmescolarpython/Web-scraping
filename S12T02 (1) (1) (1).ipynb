```python
#Llibreries import
import requests
import pandas as pd
import re
```

```python
#Llibreries from
from bs4 import BeautifulSoup #BeautifulSoup 4
from pprint import pprint
```

```python
#Accedim al website de la bolsa de Madrid.
URL = "https://www.bolsamadrid.es"
web = requests.get(URL)
print(web.text)#Mostrem el contingut de bolsamadrid.es
```

```python
#BeautifulSoup - 1
```

```python
bsoup = BeautifulSoup("Bolsa de Madrid: Notas de Prensa")
print(bsoup.prettify())##Visualització en format standard senzill.
```

```python
pprint(bsoup.prettify())#Visualització en format standard pretty.
```

```python
html = BeautifulSoup(web.text, "html.parser")
```

```python
webs = html.find(id="ctl00_webs")#Recerca de dades al web bolsamadrid.es.
```

```python
print(webs)#Relació de links a websites existents al web bolsamadrid.es.
```

```python
list = [] 

for i in webs: #Netegem els elements html.
    line = str(i)
    line = line.replace('<option value="',"")
    line = line.replace('</option>',"")
    line = line.replace('">'," ")
    
    if i.find("www") != -1:
        list.append(line)
        
for j in list:
    print(j)#Visualitzem el llistat resultant de links.
```

```python
#Accedim al website de la bolsa de Madrid.
#URL = "https://www.bolsamadrid.es/esp/aspx/ETFs/Mercados/Precios.aspx"
URL = "https://www.bolsamadrid.es/esp/aspx/ETFs/Mercados/Dividendos.aspx"
web = requests.get(URL)
soup = BeautifulSoup(web.text, "html.parser")
```

```python
soup
```

```python
#Trobem aquesta linia de codi html a la pàgina web bolsademadrid.
#<table cellpadding="3" cellspacing="0" class="TblPort" id="ctl00_Contenido_tblETFs" width="100%">
dades = soup.find(class_="TblPort", id="ctl00_Contenido_tblDatos")
dades.text
```

```python
columnes = dades.text[:72]
```

```python
columnes = columnes.replace('\n\n','')
```

```python
columnes = columnes.replace('FechaEx-Dividendo','FechaEx-Dividendo ')
columnes = columnes.replace('FechaAbono','FechaAbono ')
columnes = columnes.replace('Valor','Valor ')
columnes = columnes.replace('Ejercicio','Ejercicio ')
```

```python
columnes
```

```python
columnes = columnes.split(' ')
```

```python
columnes
```

```python
dades.text[72:]
```

```python
info = dades.text
```

```python
valors = info[72:] #Recollim els valors.
valors = valors.replace(', ',' ')
```

```python
valors = valors.replace('2022','2022,')#La coma es posa com a separador de columnes.
valors = valors.replace('2021','2021,')
```

```python
valors = valors.replace('Ord.','Ord.,')
```

```python
valors = valors.replace('A cuenta',',A cuenta')
```

```python
valors = valors.replace('.A cuenta',',A cuenta')
```

```python
valors = valors.replace('COTIZADO ARM','COTIZADO ARM,')#Netegem els literals de comes. Només posem coma com a separador.
valors = valors.replace('ETF,FI COTIZ,ARM.','ETF FI COTIZ ARM,')
valors = valors.replace('ETF ACC','ETF ACC,')
```

```python
valors = valors.replace('0,','0.')#El valor númeric millor amb punt.
```

```python
valors
```

```python
separator = "\n\n"
valors = valors.split(separator)
print(valors)
```

```python
valors.remove('')#Esborrem el darrer element de valors(es buit).
```

```python
print(columnes)#Dades de DIVIDENDOS.
for i in valors:
    item = i.replace(',','\t')
    print(item)
```

```python
import csv

df = pd.DataFrame(valors) 
df.to_csv('Dividendos.csv') #Desem el contingut en un fitxer 'Dividendos' en format csv.
```
